```python
# Treemap Visualization

import matplotlib.pyplot as plt
import squarify
import matplotlib.colors as mcolors
import matplotlib.patches as mpatches

# Data
categories = ['CNN', 'Deep CNN', 'Random Forest', 'SVM', 'K-Means', 'SOM', 'CNN Autoencoder']
values = [50, 30, 10, 20, 15, 25, 5]

# Sort the data in descending order
sorted_indices = sorted(range(len(values)), key=lambda i: values[i], reverse=True)
categories = [categories[i] for i in sorted_indices]
values = [values[i] for i in sorted_indices]

# Calculate percentages for labels
total = sum(values)
percentages = [f"{(value / total) * 100:.1f}%" for value in values]

# Combine categories with percentages
labels = [f"{cat}\n{perc}" for cat, perc in zip(categories, percentages)]

# Color Palette
# Use a colormap to generate distinct colors
cmap = plt.cm.get_cmap('Set3', len(categories))
colors = [cmap(i) for i in range(len(categories))]

# Create the treemap
plt.figure(figsize=(14, 9))
ax = plt.gca()
squarify.plot(
    sizes=values,
    label=labels,
    color=colors,
    alpha=0.9,
    edgecolor='white',
    linewidth=2,
    ax=ax
)

# Title with enhanced styling
plt.title(
    'Top Seven Most Commonly Used Machine Learning Methodologies in Radio Astronomy',
    fontsize=18,
    fontweight='bold',
    pad=20
)

# Remove axes
plt.axis('off')

# Add annotations or subtitle if needed
# plt.text(0.5, 0.95, 'Data Source: Your Data Source', horizontalalignment='center', fontsize=12, transform=ax.transAxes)

# Add a legend (optional)
# Create custom patches for the legend
patches = [mpatches.Patch(color=colors[i], label=f"{categories[i]} ({percentages[i]})") for i in range(len(categories))]
plt.legend(
    handles=patches,
    loc='upper left',
    bbox_to_anchor=(1, 1),
    fontsize=12,
    title="Categories",
    title_fontsize=14,
    frameon=False
)

# Adjust layout for better spacing
plt.tight_layout()

# Add a subtle background color
plt.gca().set_facecolor('#f9f9f9')

# Display the treemap
plt.show()
```

```python
# Treemap Visualization

import matplotlib.pyplot as plt
import squarify
import matplotlib.colors as mcolors

# Data
categories = ['CNN', 'Deep CNN', 'Random Forest', 'SVM', 'K-Means', 'SOM', 'CNN Autoencoder']
values = [50, 30, 10, 20, 15, 25, 5]

# Calculate percentages for labels
total = sum(values)
percentages = [f"{(value / total) * 100:.1f}%" for value in values]

# Combine categories with percentages
labels = [f"{cat}\n{perc}" for cat, perc in zip(categories, percentages)]

# Color Palette
# Use a colormap to generate distinct colors
cmap = plt.cm.get_cmap('Set3', len(categories))
colors = [cmap(i) for i in range(len(categories))]

# Create the treemap
plt.figure(figsize=(12, 8))
squarify.plot(
    sizes=values,
    label=labels,
    color=colors,
    alpha=0.8,
    edgecolor='white',
    linewidth=2
)

# Title with enhanced styling
plt.title(
    'Top Seven Most Commonly Used Machine Learning Methodologies in Radio Astronomy',
    fontsize=16,
    fontweight='bold'
)

# Remove axes
plt.axis('off')

# Add a legend (optional)
# Create custom patches for the legend
import matplotlib.patches as mpatches

patches = [mpatches.Patch(color=colors[i], label=f"{categories[i]} ({percentages[i]})") for i in range(len(categories))]
plt.legend(
    handles=patches,
    loc='upper right',
    bbox_to_anchor=(1.15, 1),
    fontsize=12,
    title="Categories",
    title_fontsize=14
)

# Adjust layout for better spacing
plt.tight_layout()

# Display the treemap
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import matplotlib.cm as cm
import numpy as np

# ================================================
# Step 1: Import Required Libraries
# ================================================

import matplotlib.pyplot as plt

# ================================================
# Step 2: Define the Data
# ================================================

# Categories representing different machine learning methodologies
categories = ['CNN', 'Deep CNN', 'Random Forest', 'SVM', 'K-Means', 'SOM', 'CNN Autoencoder']

# Corresponding values representing the number of papers
values = [50, 30, 10, 20, 15, 25, 5]

# ================================================
# Step 3: Sort the Data
# ================================================

# Sort the categories and values in descending order based on values
sorted_indices = np.argsort(values)[::-1]
sorted_categories = [categories[i] for i in sorted_indices]
sorted_values = [values[i] for i in sorted_indices]

# Update variables
categories = sorted_categories
values = sorted_values

# ================================================
# Step 4: Calculate Bubble Sizes and Colors
# ================================================

# Normalize the sizes for better visualization
min_size = 300    # Minimum bubble size
max_size = 1500   # Maximum bubble size
sizes = np.linspace(min_size, max_size, len(values))

# Alternatively, scale sizes based on values
sizes = [v * 30 for v in values]  # Adjust the multiplier as needed

# Choose a color map
cmap = cm.viridis

# Normalize the color mapping based on values
norm = mcolors.Normalize(vmin=min(values), vmax=max(values))
colors = cmap(norm(values))

# ================================================
# Step 5: Assign Y-Positions to Prevent Overlap
# ================================================

# Assign y positions with some offset to spread the bubbles vertically
y_positions = np.arange(len(categories))  # e.g., 0,1,2,...
y = y_positions

# Alternatively, add slight random jitter to y positions
# np.random.seed(42)  # For reproducibility
# y = y_positions + np.random.uniform(-0.1, 0.1, size=len(categories))

# ================================================
# Step 6: Create the Bubble Chart
# ================================================

fig, ax = plt.subplots(figsize=(12, 8))

# Create scatter plot
scatter = ax.scatter(categories, y, s=sizes, c=colors, alpha=0.6, edgecolors="w", linewidth=2, cmap='viridis')

# ================================================
# Step 7: Add Labels to Bubbles
# ================================================

for i, (category, value) in enumerate(zip(categories, values)):
    ax.text(category, y[i], str(value), fontsize=12, ha='center', va='center', color='white', fontweight='bold')

# ================================================
# Step 8: Customize Axes
# ================================================

# Remove y-axis as it's not meaningful in this context
ax.get_yaxis().set_visible(False)

# Set title and labels
ax.set_title('Top Seven Most Commonly Used Machine Learning Methodologies in Radio Astronomy', fontsize=16, pad=20)
ax.set_xlabel('Machine Learning Methodologies', fontsize=14)
# Optionally, you can add a dummy y-label or leave it out

# Add gridlines for better readability
ax.grid(axis='x', linestyle='--', alpha=0.7)

# ================================================
# Step 9: Add Colorbar
# ================================================

# Create a scalar mappable for the colorbar
sm = cm.ScalarMappable(cmap=cmap, norm=norm)
sm.set_array([])  # Only needed for older versions of matplotlib

# Add colorbar to the plot
cbar = plt.colorbar(sm, ax=ax, orientation='vertical', pad=0.02)
cbar.set_label('Number of Papers', fontsize=12)

# ================================================
# Step 10: Enhance Aesthetics
# ================================================

# Improve layout
plt.tight_layout()

# ================================================
# Step 11: Display the Plot
# ================================================

plt.show()
```
