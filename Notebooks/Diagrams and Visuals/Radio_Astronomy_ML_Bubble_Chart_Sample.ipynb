```python
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import matplotlib.cm as cm
import numpy as np

# ================================================
# Step 1: Import Required Libraries
# ================================================

import matplotlib.pyplot as plt

# ================================================
# Step 2: Define the Data
# ================================================

# Categories representing different machine learning methodologies
categories = ['CNN', 'Deep CNN', 'Random Forest', 'SVM', 'K-Means', 'SOM', 'CNN Autoencoder']

# Corresponding values representing the number of papers
values = [50, 30, 10, 20, 15, 25, 5]

# ================================================
# Step 3: Sort the Data
# ================================================

# Sort the categories and values in descending order based on values
sorted_indices = np.argsort(values)[::-1]
sorted_categories = [categories[i] for i in sorted_indices]
sorted_values = [values[i] for i in sorted_indices]

# Update variables
categories = sorted_categories
values = sorted_values

# ================================================
# Step 4: Calculate Bubble Sizes and Colors
# ================================================

# Normalize the sizes for better visualization
min_size = 300    # Minimum bubble size
max_size = 1500   # Maximum bubble size
sizes = np.linspace(min_size, max_size, len(values))

# Alternatively, scale sizes based on values
sizes = [v * 30 for v in values]  # Adjust the multiplier as needed

# Choose a color map
cmap = cm.viridis

# Normalize the color mapping based on values
norm = mcolors.Normalize(vmin=min(values), vmax=max(values))
colors = cmap(norm(values))

# ================================================
# Step 5: Assign Y-Positions to Prevent Overlap
# ================================================

# Assign y positions with some offset to spread the bubbles vertically
y_positions = np.arange(len(categories))  # e.g., 0,1,2,...
y = y_positions

# Alternatively, add slight random jitter to y positions
# np.random.seed(42)  # For reproducibility
# y = y_positions + np.random.uniform(-0.1, 0.1, size=len(categories))

# ================================================
# Step 6: Create the Bubble Chart
# ================================================

fig, ax = plt.subplots(figsize=(12, 8))

# Create scatter plot
scatter = ax.scatter(categories, y, s=sizes, c=colors, alpha=0.6, edgecolors="w", linewidth=2, cmap='viridis')

# ================================================
# Step 7: Add Labels to Bubbles
# ================================================

for i, (category, value) in enumerate(zip(categories, values)):
    ax.text(category, y[i], str(value), fontsize=12, ha='center', va='center', color='white', fontweight='bold')

# ================================================
# Step 8: Customize Axes
# ================================================

# Remove y-axis as it's not meaningful in this context
ax.get_yaxis().set_visible(False)

# Set title and labels
ax.set_title('Top Seven Most Commonly Used Machine Learning Methodologies in Radio Astronomy', fontsize=16, pad=20)
ax.set_xlabel('Machine Learning Methodologies', fontsize=14)
# Optionally, you can add a dummy y-label or leave it out

# Add gridlines for better readability
ax.grid(axis='x', linestyle='--', alpha=0.7)

# ================================================
# Step 9: Add Colorbar
# ================================================

# Create a scalar mappable for the colorbar
sm = cm.ScalarMappable(cmap=cmap, norm=norm)
sm.set_array([])  # Only needed for older versions of matplotlib

# Add colorbar to the plot
cbar = plt.colorbar(sm, ax=ax, orientation='vertical', pad=0.02)
cbar.set_label('Number of Papers', fontsize=12)

# ================================================
# Step 10: Enhance Aesthetics
# ================================================

# Improve layout
plt.tight_layout()

# ================================================
# Step 11: Display the Plot
# ================================================

plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
categories = ['CNN', 'Deep CNN', 'Random Forest', 'SVM', 'K-Means', 'SOM', 'CNN Autoencoder']
values = [50, 30, 10, 20, 15, 25, 5]
x = range(len(categories))
y = [1] * len(categories)
sizes = [v * 100 for v in values]  # Adjust size for better visibility

# Create the bubble chart
fig, ax = plt.subplots(figsize=(10, 6))
scatter = ax.scatter(x, y, s=sizes, c=sizes, alpha=0.5, cmap='viridis')
ax.set_xticks(x)
ax.set_xticklabels(categories)
ax.set_yticks([])
ax.set_title('Top Seven Most Commonly Used Machine Learning Methodologies in Radio Astronomy')
plt.colorbar(scatter, ax=ax, label='Number of Papers')
plt.show()

#I like this idea (but might not use it)
```
