```python
import pandas as pd
from google.colab import files

# Upload the file
uploaded = files.upload()

# Assuming the file is uploaded, you can get the filename
for filename in uploaded.keys():
    print(f"Processing file: {filename}")

# Load the dataset (replace the filename if necessary)
df = pd.read_excel(filename)

# Define relevant keywords for filtering
keywords = [
    'RFI', 'mitigation', 'interference excision', 'time-frequency analysis', 'RFI Mitigation Strategies',
    'automated RFI detection', 'automated', 'supervised learning', 'unsupervised learning', 'reinforcement learning',
    'data denoising', 'radio astronomy RFI mitigation', 'radio astronomy', 'interference', 'Interference Cancellation',
    'signal processing', 'artificial intelligence', 'radio frequency interference', 'Thresholding Techniques',
    'pulsar', 'fast radio bursts', 'FRB', 'RFI flagging', 'RFI detection', 'RFI Shielding', 'Interference Modeling',
    'radio frequency interference detection', 'Autoencoders', 'radio telescope', 'fast radio transients',
    'transient detection', 'Machine Learning', 'Deep Learning', 'Neural Network'
]

# Function to filter rows containing the keywords in specified columns and having a journal name
def filter_papers(row, columns):
    # Check if 'Journal Name' column has a non-empty value
    journal_name = row.get('Journal Name', '')
    if pd.isna(journal_name) or str(journal_name).strip() == '':
        return False

    # Join all specified column values into a single text, safely converting non-string values
    text = ' '.join([
        str(row[col]) if col in row and pd.notna(row[col]) else ''
        for col in columns
    ])

    # Check if any keyword is present in the text (case-insensitive)
    text_lower = text.lower()
    for keyword in keywords:
        if keyword.lower() in text_lower:
            return True
    return False

# Specify the columns you want to search for keywords in
columns_to_search = ['Title', 'Link']  # Replace with actual column names from your dataset

# Apply the filter
filtered_df = df[df.apply(filter_papers, columns=columns_to_search, axis=1)]

# Save the filtered dataset to a new Excel file
output_filename = 'filtered_rfi_papers.xlsx'
filtered_df.to_excel(output_filename, index=False)

# Download the filtered dataset
files.download(output_filename)

print(f"Filtered dataset saved and ready for download with {filtered_df.shape[0]} relevant papers.")
```
